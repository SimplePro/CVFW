```python
tomato_list = [f"Tomatos/tomato{i}.png" for i in range(1, 7)]
for i in range(len(tomato_list)):
    s = cv2.imread(tomato_list[i], cv2.IMREAD_COLOR)
    g = cv2.resize(cv2.cvtColor(s, cv2.COLOR_BGR2GRAY), dsize = (256, 256))
    plt.subplot(2, 3, i+1)
    plt.imshow(g, cmap="gray")
```

```python
plt.figure(figsize = (20, 2))
plt.imshow([[i for i in range(1, 255)]], cmap="gray")
```

```python
tomato = []

for file_name in tomato_list:
    s = cv2.imread(file_name, cv2.IMREAD_COLOR)
    one_d = []
    g = cv2.resize(cv2.cvtColor(s, cv2.COLOR_BGR2GRAY), dsize=(256, 256))
    for j in g.tolist():
        for k in j:
            one_d.append(k)
    tomato.append(one_d)
    
print(tomato)
```

```python
p = []
max_p, min_p, avg_p = [], [], []  # max(p_i), min(p_i), avg(p_i)

for i in range(256*256):  # p_i 번째.
    p_ = []
    for j in range(6):
        p_.append(tomato[j][i])
    p.append(p_)
    
print(len(p), len(p[0]))
```

```python
max_min = []  # max(p_i) - min(p_i)

for i in range(256*256):
    max_p.append(max(p[i]))
    min_p.append(min(p[i]))
    avg_p.append(sum(p[i]) / len(p[i]))
    max_min.append(max_p[i] - min_p[i])

    
print(sorted(max_min[:3]))
```

```python
avg_p
```

```python
a = sorted(max_min[:3])  # max - min 의 값이 가장 작은 3개

important_p = []

for i in range(256*256):
    if max_min[i] in a:
        print(max_min[i])
        important_p.append(i)

print(important_p)
```

```python
len(important_p)
```

```python
for i in range(6):
    s = cv2.imread(tomato_list[i], cv2.IMREAD_COLOR)
    g = cv2.resize(cv2.cvtColor(s, cv2.COLOR_BGR2GRAY), dsize= (256, 256))
    for j in important_p:
        g[j//256][j%256] = 255
    
    plt.subplot(2, 3, i+1)
    plt.imshow(g, cmap="gray")
```

```python
from time import sleep

src = cv2.imread("Tomatos/tomatos.jpg", cv2.IMREAD_COLOR)
gray = cv2.resize(cv2.cvtColor(src, cv2.COLOR_BGR2GRAY), dsize = (1200, 400))
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')

for i in range(400):
    for j in range(1200):
        for k in important_p:
            try:
                if abs(gray[i][j] - avg_p[k]) <= 27: gray[i][j] = 255
            except: print(i, j)

plt.subplot(1, 2, 2)
plt.imshow(gray, cmap='gray')
```
