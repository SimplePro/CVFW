```python
from cvfwVer2 import CVFW_MODEL, CVFW_UPDATE
```

```python
hand_model = CVFW_MODEL(dsize=(100, 100))

hand_model.add_directory(class_name="rock", path="C:\\kimdonghwan\\python\\CVFW\\image\\train\\rock_scissors_paper\\rock")
hand_model.add_directory(class_name="scissors", path="C:\\kimdonghwan\\python\\CVFW\\image\\train\\rock_scissors_paper\\scissors")
hand_model.add_directory(class_name="paper", path="C:\\kimdonghwan\\python\\CVFW\\image\\train\\rock_scissors_paper\\paper")
```

```python
hand_model.train()
```

```python
hand_update = CVFW_UPDATE(hand_model, feature_group_number = [10], feature_weight_number = [18])
hand_update.add_validation(class_name="rock", path="C:\\kimdonghwan\\python\\CVFW\\image\\test\\rock_scissors_paper\\rock")
hand_update.add_validation(class_name="scissors", path="C:\\kimdonghwan\\python\\CVFW\\image\\test\\rock_scissors_paper\\scissors")
hand_update.add_validation(class_name="paper", path="C:\\kimdonghwan\\python\\CVFW\\image\\test\\rock_scissors_paper\\paper")
hand_update.update()
```

```python
hand_update.set(feature_group_number = [150], feature_weight_number = [0, 50] )
hand_update.update()
```

```python
model = CVFW_MODEL(dsize=(100, 100), feature_group_number=150, feature_weight_number = 0)
model.add_directory(class_name="rock", path="C:\\kimdonghwan\\python\\CVFW\\image\\train\\rock_scissors_paper\\rock")
model.add_directory(class_name="scissors", path="C:\\kimdonghwan\\python\\CVFW\\image\\train\\rock_scissors_paper\\scissors")
model.add_directory(class_name="paper", path="C:\\kimdonghwan\\python\\CVFW\\image\\train\\rock_scissors_paper\\paper")
model.train()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.cvtColor(cv2.imread("C:\\kimdonghwan\\python\\CVFW\\image\\test\\rock_scissors_paper\\rock\\ (601).png", 1), cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap="gray")
print(model.predict_class(img.flatten().tolist()))
```
